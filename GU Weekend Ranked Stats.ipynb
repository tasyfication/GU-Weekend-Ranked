# GU Weekend Ranked Stats Tracker

```python
import pandas as pd
import requests
import json
from datetime import datetime, timedelta

base_url = "https://api.godsunchained.com/v0/match?"
user_id = input("Enter your user ID: ")  # e.g. 1831645
start_time = input("Enter start time: ") # e.g. 1640350800
end_time = input("Enter end time: ")     # e.g. 1640604600
won_url = "%splayer_won=%s&page=1&perPage=100&start_time=%s-%s" % (base_url, user_id, start_time, end_time)
lost_url = "%splayer_lost=%s&page=1&perPage=100&start_time=%s-%s" % (base_url, user_id, start_time, end_time)
won, lost = 0, 0
match_records = []
won_records = requests.get(won_url).json().get("records")
lost_records = requests.get(lost_url).json().get("records")
match_records = won_records + lost_records
match_records.sort(key = lambda x: x.get("start_time"))
match_list = []
count = 1

for x in match_records:
  x["end_time"] = (datetime.fromtimestamp(x["end_time"]) + timedelta(hours=8)).strftime("%d/%m/%Y - %-I:%M%p")
  if(x["total_rounds"] > 4 and x["game_mode"] == 13):
    if(x["player_won"] == int(user_id)):
      match_list.append([count, x["end_time"], x["player_info"][0]["god"], x["player_info"][1]["god"], "Win", x["game_id"]])
      won += 1
    else:
      match_list.append([count, x["end_time"], x["player_info"][1]["god"], x["player_info"][0]["god"], "Lost", x["game_id"]])
      lost += 1
    if(count == 25):
      match_df = pd.DataFrame(match_list, columns = ["Game #", "Date & Time", "Your God", "Opponent's God", "Result", "Game ID"])
      match_df.set_index(["Game #", "Date & Time", "Your God", "Opponent's God", "Result", "Game ID"], inplace=True)
      display(match_df)
      print("Weekend Event - Win: %s | Lost: %s | Total: %s" % (won, lost, count))
      match_df = match_df.iloc[0:0]
      match_list.clear()

    count += 1

match_df = pd.DataFrame(match_list, columns = ["Game #", "Date & Time", "Your God", "Opponent's God", "Result", "Game ID"])
match_df.set_index(["Game #", "Date & Time", "Your God", "Opponent's God", "Result", "Game ID"], inplace=True)
display(match_df)
print("Total - Win: %s | Lost: %s | Total: %s" % (won, lost, count-1))
```
